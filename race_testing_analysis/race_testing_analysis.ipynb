```python
def get_num_words(desc: str):
    """
    Gets the number of tokens (words) from a given description string. cleans punctuation before processing
    """

    clean_description = desc.replace(",", " ")
    clean_description_li = clean_description.split(" ")
    tokens = [word for word in  clean_description_li if word != ""]
    num_tokens = len(tokens)
    return num_tokens
```
